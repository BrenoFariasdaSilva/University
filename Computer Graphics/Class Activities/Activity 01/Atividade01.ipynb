```python
from PIL import Image
import requests
from io import BytesIO
from IPython.display import SVG,display
```

```python
url_imagem = 'https://upload.wikimedia.org/wikipedia/commons/thumb/3/35/Tux.svg/800px-Tux.svg.png'

# Lendo a imagem
response = requests.get(url_imagem)
img = Image.open(BytesIO(response.content))

# Exibindo a imagem
display(img)
```

```python
# Obtem informações da imagem
largura = img.size[0]
altura = img.size[1]

# Aumenta em 10x o seu tamanho
img = img.resize((largura*10, altura*10))

# Exibe a imagem com o tamanho ampliado
display(img)
```
